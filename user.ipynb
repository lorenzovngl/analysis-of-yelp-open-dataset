```python
import json 
import numpy as np
import pandas as pd

data_size = 10000
train_size = int(data_size*0.9)


# Users
with open("yelp_dataset/yelp_academic_dataset_user.json", 'r') as f:
    business_head = [next(f) for x in range(data_size)]
f.close()

users_data = []
for line in business_head:
    users_data.append(json.loads(line))

df = pd.DataFrame(users_data)

print(df.columns)
print(df.size)
```
